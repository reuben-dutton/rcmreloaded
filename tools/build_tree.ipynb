```python
import json
import pickle

import scipy
```

```python
JSON_PATH = '../data/colors.json'
TREE_PATH = '../data/tree.pickle'
```

```python
with open(JSON_PATH, 'r') as j:
    colors = json.load(j)
```

```python
labels = []
points = []

for name, rgb in colors.items():
    labels.append(name)
    points.append(rgb)

tree = scipy.spatial.KDTree(points)
```

```python
with open(TREE_PATH, 'wb') as f:
	pickle.dump((tree, labels), f)
```
