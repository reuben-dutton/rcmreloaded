```python
import os
import sys

import scipy.spatial

# allow importing the project's db package when run from tools/
sys.path.insert(0, os.path.abspath('..'))

from db import session_scope, all_colours, save_tree
```

```python
# colours are the source of truth in the database now
with session_scope() as session:
    colours = all_colours(session)
```

```python
labels = [c.name for c in colours]
points = [c.rgb for c in colours]
```

```python
tree = scipy.spatial.KDTree(points)
```

```python
# store the (tree, labels) pickle as a DB artifact (replaces data/tree.pickle)
save_tree(tree, labels)
print(f'Stored colour tree over {len(labels)} colours in the database.')
```
